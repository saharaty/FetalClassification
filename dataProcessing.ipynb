# Inspecting db

```python
import pandas as pd

# Load CSV with the correct delimiter
csv_path = '/content/drive/MyDrive/project/FETAL_PLANES_DB_data.csv'
labels_df = pd.read_csv(csv_path, delimiter=';')

# Check the column names and the first few rows
labels_df.head(), labels_df.columns
```

```python
labels_df.columns = labels_df.columns.str.strip()
```

```python
labels_df.isnull().sum()
```

```python
labels_df['Train'].unique()
```

```python
# Get all image filenames from the dataset folder
image_filenames = os.listdir(images_path)

# Check if every image in the folder matches a row in the CSV
missing_files = [img for img in image_filenames if img not in labels_df['Image_name'].values]

print(f"Missing files (if any): {missing_files}")

# Check if the CSV contains any images not in the dataset folder
missing_in_csv = [img for img in labels_df['Image_name'].values if img not in image_filenames]
print(f"Missing in CSV: {missing_in_csv}")
```

```python
# Clean image filenames by removing the extension
image_filenames_clean = [img.split('.')[0] for img in image_filenames]

# Compare cleaned filenames
missing_files_count = len([img for img in image_filenames_clean if img not in labels_df['Image_name'].values])
missing_in_csv_count = len([img for img in labels_df['Image_name'].values if img not in image_filenames_clean])

# Output the counts
print(f"Number of files missing in CSV: {missing_in_csv_count}")
print(f"Number of files missing in the dataset folder: {missing_files_count}")
```

```python
# Remove rows in the CSV where the image doesn't exist in the dataset folder
valid_rows = labels_df[labels_df['Image_name'].isin(image_filenames_clean)]

# Display the cleaned dataset
valid_rows.head()
```

```python
# Check for missing images in the dataset folder (compared to valid rows in CSV)
missing_in_images = [img for img in image_filenames_clean if img not in valid_rows['Image_name'].values]

# Check for missing rows in the CSV (compared to the image filenames)
missing_in_csv = [img for img in valid_rows['Image_name'].values if img not in image_filenames_clean]

# Output the number of missing files in each
print(f"Number of images missing in valid CSV: {len(missing_in_csv)}")
print(f"Number of images missing in dataset folder: {len(missing_in_images)}")
```

```python
# Group the images by their 'Plane' label
grouped_by_plane = valid_rows.groupby('Plane')['Image_name'].apply(list)

# Check how the images are distributed by plane
print(grouped_by_plane)
```

```python
from sklearn.model_selection import train_test_split

# Create an empty split column
valid_rows['split'] = ''

# Go plane by plane
for plane, group in valid_rows.groupby('Plane'):
    train, temp = train_test_split(group, test_size=0.3, random_state=42)
    val, test = train_test_split(temp, test_size=0.5, random_state=42)

    valid_rows.loc[train.index, 'split'] = 'train'
    valid_rows.loc[val.index, 'split'] = 'val'
    valid_rows.loc[test.index, 'split'] = 'test'
```

```python
print(valid_rows.groupby(['Plane', 'split']).size())
```

```python
import os

base_dir = '/content/final_dataset'
images_path = '/content/drive/MyDrive/project/Images'

# Create destination folders
for split in ['train', 'val', 'test']:
    for plane in valid_rows['Plane'].unique():
        os.makedirs(os.path.join(base_dir, split, plane), exist_ok=True)

# Create a shell script with rsync commands
with open("copy_images.sh", "w") as f:
    for _, row in valid_rows.iterrows():
        img_name = row['Image_name'] + '.png'
        plane = row['Plane']
        split = row['split']

        src = os.path.join(images_path, img_name)
        dst = os.path.join(base_dir, split, plane, img_name)

        if os.path.exists(src):
            f.write(f'rsync -a "{src}" "{dst}"\n')
```
